```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/umanggoel/Downloads/Salary.csv")
```

```python
X = data["Area(in acres)"].values.reshape(-1, 1)
y = data["Price"].values
```

```python
X_mean = np.mean(X)
X_std = np.std(X)
X = (X-X_mean)/X_std
```

```python
X = np.c_[np.ones(X.shape[0]), X]
```

```python
theta = np.zeros(X.shape[1])
```

```python
alpha = 0.01
num_iterations = 1000
```

```python
for i in range(num_iterations):
    # Calculate the predicted values
    y_pred = X.dot(theta)
    
    # Calculate the error between the predicted values and the true values
    error = y_pred - y
    
    # Update the coefficients using the LMS algorithm
    theta -= alpha * X.T.dot(error) / X.shape[0]    #(coeff = (x-X_)*(ypred-y)/(x-X_)^2)
    
```

```python
print(theta)
```

```python
plt.scatter(X[:, 1], y, alpha=0.7)
plt.plot(X[:, 1], X.dot(theta), color='red')
# plt.plot(X,y,color='green')
plt.xlabel("Area (acres)")
plt.ylabel("Price ($)")
plt.show()
```
